```python
for i in merged_df.index:
    if merged_df.loc[i, "buyer_country"] == merged_df.loc[i, "seller_country"]:
        merged_df.loc[i, "locality"] = "domestic"
    else:
        merged_df.loc[i, "locality"] = "international"
```

```python
merged_df
```

```python
df1 = merged_df.groupby(["locality", "brand"]).mean("total_price").loc["domestic"]
df1
```

```python
df2 = merged_df.groupby(["locality", "brand"]).mean("total_price").loc["international"]
df2
```
