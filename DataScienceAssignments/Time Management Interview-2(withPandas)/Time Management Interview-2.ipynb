2. Which day has the highest value of purchases?

7. Which brand had the largest absolute â¬ difference in average transaction value between domestic and international?

```python
merged_df.head()
```

```python
for i in merged_df.index:
    if merged_df.loc[i, "buyer_country"] == merged_df.loc[i, "seller_country"]:
        merged_df.loc[i, "total_domestic"] = "domestic"
    else:
        merged_df.loc[i, "total_domestic"] = "international"
```

```python
merged_df
```

```python
df1 = merged_df.groupby(["total_domestic", "brand"]).mean("total_rate").loc["domestic"]
```

```python
df2 = merged_df.groupby(["total_domestic", "brand"]).mean("total_rate").loc["international"]
```

```python
df1.reset_index(inplace=True)
```

```python
df2.reset_index(inplace=True)
```

```python
df_last = pd.merge(df1, df2, on=["brand"])
```

```python
df_last
```
